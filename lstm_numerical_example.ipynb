```python
import numpy as np

class LSTMNetwork:
    def __init__(self):
        self.W_xf = 0.5
        self.W_hf = 0.1
        self.b_f = 0
        
        self.W_xi = 0.6
        self.W_hi = 0.2
        self.b_i = 0
        
        self.W_xc = 0.7
        self.W_hc = 0.3
        self.b_c = 0
        
        self.W_xo = 0.8
        self.W_ho = 0.4
        self.b_o = 0
        
        self.W_y = 4
        self.b_y = 0

    def sigmoid_activation(self, value):
        return 1 / (1 + np.exp(-value))

    def tanh_activation(self, value):
        return np.tanh(value)

    def compute_forget_gate(self, x_t, h_prev):
        net_f = (self.W_xf * x_t) + (self.W_hf * h_prev) + self.b_f
        return self.sigmoid_activation(net_f)

    def compute_input_gate(self, x_t, h_prev):
        net_i = (self.W_xi * x_t) + (self.W_hi * h_prev) + self.b_i
        return self.sigmoid_activation(net_i)

    def compute_candidate_state(self, x_t, h_prev):
        net_c = (self.W_xc * x_t) + (self.W_hc * h_prev) + self.b_c
        return self.tanh_activation(net_c)

    def compute_output_gate(self, x_t, h_prev):
        net_o = (self.W_xo * x_t) + (self.W_ho * h_prev) + self.b_o
        return self.sigmoid_activation(net_o)

    def update_cell_state(self, f_t, c_prev, i_t, c_tilde_t):
        return (f_t * c_prev) + (i_t * c_tilde_t)

    def update_hidden_state(self, o_t, c_t):
        return o_t * self.tanh_activation(c_t)

    def linear_output(self, h_t):
        return (self.W_y * h_t) + self.b_y

def run_numerical_simulation():
    lstm = LSTMNetwork()
    
    current_h = 0
    current_c = 0
    
    input_sequence = [1, 2, 3]
    prediction_target = 4
    
    print("="*50)
    print("LSTM FROM SCRATCH - NUMERICAL EXAMPLE")
    print("="*50)
    
    for t, x_t in enumerate(input_sequence, start=1):
        f_t = lstm.compute_forget_gate(x_t, current_h)
        i_t = lstm.compute_input_gate(x_t, current_h)
        c_tilde_t = lstm.compute_candidate_state(x_t, current_h)
        
        new_c = lstm.update_cell_state(f_t, current_c, i_t, c_tilde_t)
        
        o_t = lstm.compute_output_gate(x_t, current_h)
        new_h = lstm.update_hidden_state(o_t, new_c)
        
        print(f"TIME STEP t = {t} | INPUT x_t = {x_t}")
        print(f"  Forget Gate (f_{t})      : {f_t:.4f}")
        print(f"  Input Gate (i_{t})       : {i_t:.4f}")
        print(f"  Candidate (~C_{t})      : {c_tilde_t:.4f}")
        print(f"  Cell State (C_{t})       : {new_c:.4f}")
        print(f"  Output Gate (o_{t})      : {o_t:.4f}")
        print(f"  Hidden State (h_{t})     : {new_h:.4f}")
        print("-" * 30)
        
        current_h = new_h
        current_c = new_c

    print(f"PREDICTION STEP | INPUT x_t = {prediction_target}")
    
    f_4 = lstm.compute_forget_gate(prediction_target, current_h)
    i_4 = lstm.compute_input_gate(prediction_target, current_h)
    c_tilde_4 = lstm.compute_candidate_state(prediction_target, current_h)
    c_4 = lstm.update_cell_state(f_4, current_c, i_4, c_tilde_4)
    o_4 = lstm.compute_output_gate(prediction_target, current_h)
    h_4 = lstm.update_hidden_state(o_4, c_4)
    
    print(f"  Forget Gate (f_4)      : {f_4:.4f}")
    print(f"  Input Gate (i_4)       : {i_4:.4f}")
    print(f"  Candidate (~C_4)      : {c_tilde_4:.4f}")
    print(f"  Cell State (C_4)       : {c_4:.4f}")
    print(f"  Output Gate (o_4)      : {o_4:.4f}")
    print(f"  Hidden State (h_4)     : {h_4:.4f}")
    
    y_final = lstm.linear_output(h_4)
    
    print("="*50)
    print(f"FINAL PREDICTION CALCULATION")
    print(f"y_hat = ({lstm.W_y} * h_4) + {lstm.b_y}")
    print(f"y_hat = {y_final:.4f}")
    print("="*50)

if __name__ == "__main__":
    run_numerical_simulation()
```
